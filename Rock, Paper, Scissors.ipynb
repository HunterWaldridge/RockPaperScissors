```python
import random

player_score = 0
computer_score = 0
rounds = 0
tie_counter=0

while True:
    while True:
        user_play = (input("Rock(1), Paper(2), or Scissors(3)?"))
        if user_play == "1":
            user_play = "rock"
            break
        elif user_play == "2":
            user_play = "paper"
            break
        elif user_play == "3":
            user_play = "scissors"
            break
        else:
            print("ERROR: You must enter 1 for rock, 2 for paper, or 3 for scissors.")
            continue
            
    possible_play = ["rock","paper","scissors"]
    computer_play = random.choice(possible_play)
    print(f"You chose {user_play}, computer chose {computer_play}")

    if user_play == computer_play:
        print("It's a tie...")
        tie_counter=tie_counter+1
    elif user_play == "rock":
        if computer_play == "scissors":
            print("Rock smashes scissors! You win!")
            player_score = player_score+1
        else:
            print("Paper covers rock! You lose.")
            computer_score = computer_score+1
    elif user_play == "paper":
        if computer_play == "rock":
            print("Paper covers rock! You win!")
            player_score = player_score+1
        else:
            print("Scissors cuts paper! You lose.")
            computer_score = computer_score+1
    elif user_play == "scissors":
        if computer_play == "rock":
            print("Rock smashes scissors! You lose.")
            computer_score = computer_score+1
        else:
            print("Scissors cut paper! You win!")
            player_score = player_score+1
            
    while True:
        go_again = input("Another round? y = yes; n = no:")
        if go_again == "y" or go_again == "n":
            break
        if go_again != "y" and go_again != "n":
            print("ERROR: You can only enter 'y' or 'n'!")
            continue
            
    if go_again == "n":
        rounds=rounds+1
        print(f"Rounds: {rounds}")
        print(f"Ties: {tie_counter}")
        print(f"You scored: {player_score}.") 
        print(f"The computer scored: {computer_score}.")
        if computer_score>player_score:
            print("YOU LOST THE GAME. BETTER LUCK NEXT TIME")
        if player_score>computer_score:
            print("YOU WON! CONGRATULATIONS!")
        if player_score==computer_score:
            print("IT'S A TIE! TRY AGAIN")
        break

    if go_again == "y":
        rounds=rounds+1
        print(f"Round #{rounds+1}:")
        continue
    
```
